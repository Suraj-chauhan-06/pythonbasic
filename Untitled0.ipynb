5. What are dunder methods in Python ?
Answer = In Python, dunder methods (short for "double underscore methods") are special methods that have double underscores at both the beginning and end of their names. These methods are also known as magic methods or special methods. They allow you to define how objects of your classes behave with certain operators or built-in functions.

For example, when you use operators like +, -, ==, etc., Python internally calls the corresponding dunder methods to perform those operations. These methods can also help you control the behavior of objects when interacting with functions like len(), str(), etc.

Here are a few common dunder methods:

__init__(self, ...): Constructor method for object initialization when an instance of a class is created.

__str__(self): Defines the string representation of an object (called when str() or print() is used).

__repr__(self): Defines the official string representation of an object (used by repr()).

__add__(self, other): Used to define behavior for the + operator (addition).

__eq__(self, other): Used for equality comparison (i.e., ==).

__len__(self): Defines the behavior for the len() function when called on an object.

__getitem__(self, key): Allows an object to be subscripted like a list or dictionary (e.g., obj[key]).

__setitem__(self, key, value): Allows you to set values with item assignment (e.g., obj[key] = value).

__call__(self, ...): This makes an object callable like a function. You can define what happens when you call an object (e.g., obj()).

__del__(self): Destructor method that gets called when an object is about to be destroyed.

__iter__(self): Makes an object iterable, enabling iteration using for.

__next__(self): Defines behavior for the next() function when iterating over the object.

7. What is polymorphism in OOP?

Answer = In Object-Oriented Programming (OOP), polymorphism refers to the ability of different objects to respond to the same method or message in their own way. It allows for objects of different classes to be treated as objects of a common superclass, especially when they share the same method names or interfaces.

There are two main types of polymorphism:

Compile-time polymorphism (Method Overloading):

This happens when multiple methods have the same name but differ in the number or type of their parameters.

The method to be called is resolved at compile time.

Example in Java:

java
Copy
class Calculator {
    int add(int a, int b) {
        return a + b;
    }
    
    double add(double a, double b) {
        return a + b;
    }
}
Run-time polymorphism (Method Overriding):

This happens when a subclass provides its own implementation of a method that is already defined in its superclass.

The method to be called is resolved at runtime.

Example in Java:

java
Copy
class Animal {
    void sound() {
        System.out.println("Animal makes a sound");
    }
}

class Dog extends Animal {
    @Override
    void sound() {
        System.out.println("Dog barks");
    }
}

class Main {
    public static void main(String[] args) {
        Animal myDog = new Dog();
        myDog.sound(); // Output: Dog barks
    }
}
In the above example, although myDog is of type Animal, the method sound() from the Dog class is called, because the actual object is of type Dog. This is an example of run-time polymorphism.

Benefits of Polymorphism:
Flexibility and Reusability: You can write code that works with objects of different types, allowing for more generic and reusable code.
Maintainability: It simplifies code maintenance and extensibility since you can add new subclasses without modifying existing code.
Simplifies code: It reduces the need for complex conditional statements by allowing a uniform interface for different types.

8. How is encapsulation achieved in Python

ANswer = In Python, encapsulation is achieved by restricting access to certain details of an object and allowing interaction with the object only through its methods. This is one of the key principles of object-oriented programming (OOP), which helps in protecting the internal state of an object from outside interference and misuse.

Python provides encapsulation primarily through private and public members (attributes and methods) and through the use of getters and setters. Here's how it works:

1. Public Members
By default, all members (attributes and methods) in Python are public. These can be accessed directly from outside the class.

python
Copy
class MyClass:
    def __init__(self, value):
        self.value = value  # public attribute

    def show(self):
        print(self.value)  # public method

obj = MyClass(10)
print(obj.value)  # Accessing public attribute directly
obj.show()  # Accessing public method directly
2. Private Members
To achieve encapsulation and restrict access to an object's internal data, you can make attributes and methods private by prefixing them with a double underscore (__). This name mangles the attribute or method, making it less accessible directly from outside the class.

python
Copy
class MyClass:
    def __init__(self, value):
        self.__value = value  # private attribute

    def __show(self):  # private method
        print(self.__value)

    def get_value(self):  # public method to access private attribute
        return self.__value

obj = MyClass(10)
# print(obj.__value)  # Will raise an error: AttributeError
print(obj.get_value())  # Accessing private value via a public method
Private attributes are "name-mangled" by Python to _ClassName__attribute, meaning you can still technically access them, but this is strongly discouraged.
python
Copy
print(obj._MyClass__value)  # Not recommended, but possible
3. Protected Members
By convention, members with a single underscore (_) are considered protected. These are not strictly private, but they indicate to other developers that these members should not be accessed directly from outside the class or subclass.

python
Copy
class MyClass:
    def __init__(self, value):
        self._value = value  # protected attribute

    def _show(self):  # protected method
        print(self._value)
4. Getters and Setters
Instead of directly accessing private attributes, you can use getter and setter methods to interact with them. This helps to add logic, validation, or other controls when modifying or retrieving the internal state of the object.

python
Copy
class MyClass:
    def __init__(self, value):
        self.__value = value

    def get_value(self):  # Getter method
        return self.__value

    def set_value(self, value):  # Setter method
        if value >= 0:  # Validation logic
            self.__value = value
        else:
            print("Invalid value")

obj = MyClass(10)
print(obj.get_value())  # Accessing through getter
obj.set_value(20)  # Accessing through setter
print(obj.get_value())
Key Points
Private members: Use __ (double underscore) to prevent access.
Protected members: Use _ (single underscore) to indicate that they are for internal use or subclasses.
Public members: Default behavior, accessible directly.
Getters and Setters: Used to encapsulate logic and protect the internal state.

10. What are class and static methods in Python

Answer = In Python, both class methods and static methods are methods that are bound to the class rather than instances of the class. However, they have distinct characteristics and purposes. Here's a breakdown of the differences between them:

1. Class Method
A class method is a method that is bound to the class and not the instance of the class. It takes a class as its first argument (usually named cls) rather than the instance (self).

Purpose: Class methods are used for operations that are related to the class itself, rather than an instance of the class.
How it's defined: You define a class method by using the @classmethod decorator.
Example:
python
Copy
class MyClass:
    class_variable = 0

    def __init__(self, value):
        self.value = value

    @classmethod
    def increment_class_variable(cls):
        cls.class_variable += 1
        print(f'Class variable incremented: {cls.class_variable}')

# Using the class method
MyClass.increment_class_variable()  # Class variable incremented: 1
In this example, the increment_class_variable method is defined as a class method and modifies the class-level variable class_variable.
2. Static Method
A static method is a method that does not take either self (instance) or cls (class) as its first argument. It behaves like a regular function that is part of the class, but it doesn’t interact with class or instance attributes.

Purpose: Static methods are used when you want to include a function inside the class that does not need access to class or instance-specific data.
How it's defined: You define a static method by using the @staticmethod decorator.
Example:
python
Copy
class MyClass:

    @staticmethod
    def add_numbers(a, b):
        return a + b

# Using the static method
result = MyClass.add_numbers(3, 5)
print(result)  # Output: 8
In this example, the add_numbers method doesn't access any instance or class-level data. It is simply a utility method bundled within the class.
Key Differences:
Class Method:

Takes the class (cls) as its first argument.
Can access and modify class-level variables.
Defined using the @classmethod decorator.
Static Method:

Doesn't take self or cls as arguments.
Cannot modify or access class/instance-specific attributes.
Defined using the @staticmethod decorator.

13. What is a property decorator in Python

ANswer = A property decorator in Python is used to define methods that behave like attributes. It allows you to define a method that can be accessed like an attribute, without having to explicitly call it like a method (i.e., without parentheses).

The @property decorator makes a method act as if it's a read-only attribute, enabling you to define a method that can be accessed like an attribute, but still performs some logic or computation when accessed.

Here’s an example of how it works:

Example:
python
Copy
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius
    
    @property
    def area(self):
        return 3.14159 * (self._radius ** 2)
    
    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("Radius cannot be negative")
        self._radius = value

# Usage
circle = Circle(5)
print(circle.radius)  # Accessing radius like an attribute
print(circle.area)    # Accessing area as a property
circle.radius = 10    # Setting a new radius using the setter
Explanation:
@property: Turns the radius and area methods into properties. This means they can be accessed like attributes (circle.radius) instead of being called as methods (circle.radius()).

Getter: The first method under @property is the getter method, which allows you to retrieve the value of an attribute.

Setter: The @radius.setter decorator is used to define a setter method for the radius property, so that you can set the value of radius while enforcing some validation (like ensuring it is not negative in this case).

1. __str__:
Purpose: The __str__ method is used to define a "user-friendly" or informal string representation of an object. This string is intended to be easy to read and understand when printed or converted to a string using str().
Use Case: It's typically used for display purposes, when you want to present an object to the user in a human-readable format.
Example:

Copy
Edit

2. __repr__:
Purpose: The __repr__ method is used to define an "official" or unambiguous string representation of an object. This string is meant to be more detailed and, ideally, should be valid Python code that could be used to recreate the object using eval().
Use Case: It's primarily used for debugging and logging, providing a more precise, unambiguous description of the object.

Copy
Edit
class Person:

Key Differences:
__str__: Intended for "human-readable" output.
